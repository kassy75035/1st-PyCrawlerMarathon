## 1. 取出今天所有的新聞

```python
'''
Your Code
'''
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime 
browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
from datetime import datetime 
time_now = datetime.now()
datetime.strptime(soup.find(class_="part_list_2").find_all(class_="date")[-1].text,'%Y/%m/%d %H:%M') >datetime(2020,2,3)
```

```python
browser.get("https://www.ettoday.net/news/news-list.htm")
import time
for i in range(1000):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0, 10000*%i);"%i)
    html_source = browser.page_source
    soup = BeautifulSoup(html_source, "html5lib")
    if datetime.strptime(soup.find(class_="part_list_2").find_all(class_="date")[-1].text,'%Y/%m/%d %H:%M') > datetime(2020,2,3):
        print(datetime.strptime(soup.find(class_="part_list_2").find_all(class_="date")[-1].text,'%Y/%m/%d %H:%M'))
        continue
    else:
        break

        
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
    
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
